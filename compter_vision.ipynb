```
from skimage import data
cat = data.chelsea()
plt.imshow(cat)
print(cat.shape)
print(cat.dtype)
```

```
hsv_image = cv2.cvtColor(cat,cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
```

```
import numpy as np
image = cv2.cvtColor(cat, cv2.COLOR_BGR2HSV)

increase = 100

v = image[:, :, 2]
v = np.where(v <= 255 - increase, v + increase, 255)
image[:, :, 2] = v

image = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)

plt.imshow(image)
```

```
from skimage import filters
from scipy.ndimage import convolve
square = np.zeros((10,10), dtype=float)
print (square)
```

```
square[3:7, 3:7] = 1
print (square)
```

```
fig, axes = plt.subplots()
axes.imshow(square)
```

```
mean = np.ones((4,4))/12.0
print (mean)
```

```
smooth_convolve = convolve(square, mean)
smooth = filters.gaussian(square, 1.2)
```

```
plt.imshow(smooth_convolve)
```

```
plt.imshow(smooth)
```
